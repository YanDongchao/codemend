```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
width = 0.8
lefts_a = np.arange(3)*2 + (1 - width)
lefts_b = lefts_a + width
heights_a = [2,3,4]
heights_b = [4,2,1]
plt.bar(lefts_a, heights_a, width=width, color='red', label='group A', hatch='/')
plt.bar(lefts_b, heights_b, width=width, color='blue', label='group B', hatch='\\')
plt.legend(loc='upper right', shadow=True)
plt.xlabel('foo')
plt.ylabel('bar')
plt.xticks([1,3,5],['Case 1','Case 2','Case 3'])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
plt.title('Bar plot')
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = [0.62, 0.95, 0.85]
b = [0.38, 0.62, 0.82]
data = np.array([a,b])
people = ['Adam', 'Bobby', 'Charlie']
color_map = plt.get_cmap('Pastel1', 9)
labels = ['Before', 'After']
patterns = ['/','\\']
for group in range(2):
    left = np.arange(data.shape[1])*3 + group
    height = data[group]
    color = color_map(group)
    bars = plt.bar(left, height, color=color)
    for bar in bars:
        bar.set_width(0.8)
        if bars.index(bar) == 0:
            bar.set_label(labels[group])
        bar.set_hatch(patterns[group])
plt.xlim(-1,9)
plt.xticks([1,4,7], people)
plt.legend(loc='lower center', ncol=2, shadow=True,
           bbox_to_anchor=(0.5,-.2))
plt.title('Experiment Result')
```

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels,
        colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
text = plt.title('Raining hogs and dogs', y=1.1)
text.set_bbox(dict(facecolor='lightskyblue',
                   edgecolor='grey',
                   pad=5))
plt.axis('equal')
```

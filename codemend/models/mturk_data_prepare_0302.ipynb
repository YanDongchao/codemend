## Case 1: Bar plot

### Before

```python
import numpy as np
import matplotlib.pyplot as plt
a = [0.62, 0.95, 0.85]
b = [0.38, 0.62, 0.82]
data = np.array([a,b])
people = ['Adam', 'Bobby', 'Charlie']
colors = ['red','blue']
labels = ['Case 1', 'Case 2']
patterns = ['/','\\']
for group in range(2):
    left = np.arange(data.shape[1])*3 + group
    height = data[group]
    color = colors[group]
    bars = plt.bar(left, height, color=color)
    for bar in bars:
        bar.set_width(1)
        if bars.index(bar) == 0:
            bar.set_label(labels[group])
plt.xlim(-1,9)
plt.xticks([1,4,7], people)
plt.legend(bbox_to_anchor=(1.3,0.6))
plt.title('Experiment Result');
plt.grid(True)

from matplotlib.backends.backend_pdf import PdfPages
from codemend import relative_path
pp = PdfPages(relative_path('models/output/fig-before1.pdf'))
plt.savefig(pp, format='pdf',bbox_inches='tight')
pp.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = [0.62, 0.95, 0.85]
b = [0.38, 0.62, 0.82]
data = np.array([a,b])
people = ['Adam', 'Bobby', 'Charlie']
color_map = plt.get_cmap('Pastel1', 9)
labels = ['Case 1', 'Case 2']
patterns = ['/','\\']
for group in range(2):
    left = np.arange(data.shape[1])*3 + group
    height = data[group]
    color = color_map(group)
    bars = plt.bar(left, height, color=color)
    for bar in bars:
        bar.set_width(1)
        if bars.index(bar) == 0:
            bar.set_label(labels[group])
plt.xlim(-1,9)
plt.xticks([1,4,7], people)
plt.legend(bbox_to_anchor=(1.3,0.6))
plt.title('Experiment Result');
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = [0.62, 0.95, 0.85]
b = [0.38, 0.62, 0.82]
data = np.array([a,b])
people = ['Adam', 'Bobby', 'Charlie']
color_map = plt.get_cmap('Pastel1', 9)
labels = ['Case 1', 'Case 2']
patterns = ['/','\\']
for group in range(2):
    left = np.arange(data.shape[1])*3 + group
    height = data[group]
    color = color_map(group)
    bars = plt.bar(left, height, color=color)
    for bar in bars:
        bar.set_width(0.8)
        if bars.index(bar) == 0:
            bar.set_label(labels[group])
        bar.set_hatch(patterns[group])
plt.xlim(-1,9)
plt.xticks([1,4,7], people)
plt.legend(loc='lower center', ncol=2, shadow=True, 
           bbox_to_anchor=(0.5,-.3))
plt.title('Experiment Result');


from matplotlib.backends.backend_pdf import PdfPages
from codemend import relative_path
pp = PdfPages(relative_path('models/output/fig-after1.pdf'))
plt.savefig(pp, format='pdf',bbox_inches='tight')
pp.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
np.random.seed(0)
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
arrowprops = dict(facecolor='black', shrink=0.05)
i = np.argmax(area)
arrow_end = xy=(x[i],y[i])
arrow_start = (x[i]+0.1,y[i]+0.1)
plt.annotate('biggest', arrow_end, 
             xytext=arrow_start, arrowprops=arrowprops)
plt.ylim(-0.5, 1.5)
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.grid(True)
```

## Case 5: Contour Plot

### Before

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

matplotlib.rcParams['xtick.direction'] = 'in'
matplotlib.rcParams['ytick.direction'] = 'in'

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = mlab.bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
Z2 = mlab.bivariate_normal(X, Y, 1.5, 0.5, 1, 1)
Z = 10.0 * (Z2 - Z1)

levels = np.arange(-1.2, 1.6, 0.2)
CS = plt.contour(Z, levels,
                 origin='lower',
                 linewidths=1,
                 extent=(-3, 3, -2, 2))

zc = CS.collections[6]
plt.setp(zc, linewidth=2)

plt.clabel(CS, levels[1::2],
           inline=1,
           fmt='%1.3f',
           fontsize=10)

plt.title('January 2011')
```

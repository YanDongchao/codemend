```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0.62, 0.95, 0.85])
b = np.array([0.38, 0.62, 0.82])
left = np.array([1, 4, 7])
plt.bar(left-0.8, a, width=0.8, color='red')
plt.bar(left, b, width=0.8, color='blue')
plt.xticks(left, ['A', 'B', 'C'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0.62, 0.95, 0.85])
b = np.array([0.38, 0.62, 0.82])
left = np.array([1, 4, 7])
plt.bar(left-0.8, a, width=0.8, color='red')
plt.bar(left, b, width=0.8, color='blue')
plt.xticks(left, ['A', 'B', 'C'])
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0.62, 0.95, 0.85])
b = np.array([0.38, 0.62, 0.82])
left = np.array([1, 4, 7])
plt.bar(left-0.8, a, width=0.8, color='red')
plt.bar(left, b, width=0.8, color='blue')
plt.xticks(left, ['A', 'B', 'C'], rotation=45)
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0.62, 0.95, 0.85])
b = np.array([0.38, 0.62, 0.82])
left = np.array([1, 4, 7])
plt.bar(left-0.8, a, width=0.8, color='red', hatch='*')
plt.bar(left, b, width=0.8, color='blue', hatch='/')
plt.xticks(left, ['A', 'B', 'C'], rotation=45)
plt.grid(True)
```
